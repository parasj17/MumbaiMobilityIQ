```python
import matplotlib.dates as mdates

TARGET = 'Subways: Total Estimated Ridership'

fig, axes = plt.subplots(2, 1, figsize=(15, 9))

axes[0].plot(df['Date'], df[TARGET]/1e6, color='steelblue', linewidth=0.8)
axes[0].fill_between(df['Date'], df[TARGET]/1e6, alpha=0.15, color='steelblue')
axes[0].axvline(pd.Timestamp('2020-03-22'), color='red', linestyle='--',
                linewidth=1.5, label='NYC Lockdown — Mar 22, 2020')
axes[0].set_title("MTA Subway Daily Ridership (Mar 2020 – Jan 2025)", fontweight='bold')
axes[0].set_ylabel("Ridership (Millions)")
axes[0].legend()
axes[0].xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
axes[0].xaxis.set_major_locator(mdates.MonthLocator(interval=6))
plt.setp(axes[0].xaxis.get_majorticklabels(), rotation=45)

df['rolling_30'] = df[TARGET].rolling(30, center=True).mean()
axes[1].plot(df['Date'], df[TARGET]/1e6, color='lightsteelblue', linewidth=0.5, alpha=0.5, label='Daily')
axes[1].plot(df['Date'], df['rolling_30']/1e6, color='navy', linewidth=2, label='30-Day Rolling Mean')
axes[1].set_title("Trend Extraction — 30-Day Rolling Mean", fontweight='bold')
axes[1].set_ylabel("Ridership (Millions)")
axes[1].legend()
axes[1].xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
axes[1].xaxis.set_major_locator(mdates.MonthLocator(interval=6))
plt.setp(axes[1].xaxis.get_majorticklabels(), rotation=45)

plt.tight_layout()
plt.savefig('../Reports/01_target_timeseries.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
mode_cols = [c for c in df.columns if 'Ridership' in c or 'Traffic' in c or
             'Trips' in c]
short_names = ['Subways', 'Buses', 'LIRR', 'Metro-North', 
               'Access-A-Ride', 'Bridges+Tunnels', 'Staten Island RR']

corr = df[mode_cols].corr()
corr.index   = short_names
corr.columns = short_names

plt.figure (figsize = (9, 7))
sns.heatmap (corr, annot = True, fmt = '.2f', cmap = 'coolwarm',
             center = 0, linewidths = 0.5, vmin=-1, vmax=1)
plt.title ("Cross-Modal Ridership Correlation — All MTA Transport Modes",
          fontweight='bold')
plt.tight_layout()
plt.savefig ('../Reports/01_correlation_heatmap.png', dpi=150, bbox_inches='tight')
plt.show ()

print("\nSubway correlation with other modes:")
print(corr['Subways'].drop('Subways').sort_values(ascending=False).to_string())
```
